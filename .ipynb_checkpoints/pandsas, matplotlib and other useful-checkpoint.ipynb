# FIND TOP 10 MOVIES FROM THE MOVIELENS DATASET
1. create a dataframe for movies.csv
2. create a dataframe for ratings.csv
3. Find avg rating and number of rating for each movieId based on ratings dataframe
4. Select movies that have got 100+ ratings
5. Merge with movies dataframe
6. Top 10 movies based on avg ratings
7. display movieId, title, avg_rating


```python
#loading movies and ratings dataset 
movies = pd.read_csv('C:/Users/VIBIN/Desktop/practice/data/movielens/movies.csv')
ratings = pd.read_csv('C:/Users/VIBIN/Desktop/practice/data/movielens/ratings.csv')
```

```python
movies.head()
```

```python
ratings.head()
```

```python
#find avg rating and number of rating for each movie id based on rating dataframe
ratings.groupby('movieId').rating.agg([np.mean,len]).head(10)
```

```python
#select that have 100+ rating
ratings.groupby('movieId').rating.agg([np.mean,len]).query("len > 100").head()
```

```python
agg = ratings.groupby("movieId").rating.agg([np.mean, len]).query("len >= 100").reset_index()
agg.head()
```

```python
#merge two data frames
pd.merge(movies, agg, on ='movieId').head()
```
